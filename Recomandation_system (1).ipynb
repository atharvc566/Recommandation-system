```python
user_book=book.pivot_table(index='ID',
                            columns='Title',
                              values='Rating',aggfunc='sum')
book.set_index(['ID','Title','Rating'],append=True)
```

```python
user_book.head()
```

```python
user_book.index = book.ID.unique()
```

```python
user_book
```

```python
user_book.fillna(0,inplace=True)
```

```python
user_book
```

```python
#Calculating Cosine Similarity between Users
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine
```

```python
user_sim = 1 - pairwise_distances( user_book.values,metric='cosine')
```

```python
user_sim
```

```python
#Store the results in a dataframe
user_sim_df = pd.DataFrame(user_sim)
```

```python
#Set the index and column names to user ids 
user_sim_df.index = book.ID.unique()
user_sim_df.columns = book.ID.unique()
```

```python
user_sim_df.iloc[0:5, 0:5]
```

```python
np.fill_diagonal(user_sim, 0)
user_sim_df.iloc[0:5, 0:5]
```

```python
#Most Similar Users
user_sim_df.idxmax(axis=1)[0:5]
```
